# Testing

### Función para generar dataframes.

```python
def gendf_grafico(n):
    rows = []
    file_name = 'tests/problemas/problema_%i.in' % (n)
    results_file_name = 'tests/resultados/resultado_%i.out' % (n)
    lines = [line.rstrip('\n') for line in open(file_name)]
    results_lines = [line.rstrip('\n') for line in open(results_file_name)]
    for i in range(0, len(lines)):
        datos = lines[i].split()
        if len(datos) > 1:
            x = int(datos[0])
            y = int(datos[1])
            componente = int(results_lines[i-1])
            rows.append([x, y, componente])
    return pd.DataFrame(rows, columns=['X', 'Y', 'Cluster'])
```

### Datasets pequeños. Generar dataframes y plotear.

Para los tres algoritmos: f = 1.2 y d = 1.7 producen las clusterizaciones deseadas en estos test cases.

```python
for i in range(1,5):
    df_grafico = gendf_grafico(i)
    cantidadDeComponentes = df_grafico['Cluster'].max(axis=0) + 1
    sns.scatterplot(x=df_grafico['X'], y=df_grafico['Y'], hue=df_grafico['Cluster'], palette=sns.color_palette(n_colors=cantidadDeComponentes)).legend().set_visible(False)
    plt.show()
```

### Datasets grandes. Generar dataframes y plotear.

Para los tres algoritmos: f = 3 y d = 2 producen las clusterizaciones deseadas en estos test cases.

```python
for i in range(5, 9):
    df_grafico = gendf_grafico(i)
    cantidadDeComponentes = df_grafico['Cluster'].max(axis=0) + 1
    plt.figure(figsize=(7, 7))
    sns.scatterplot(x=df_grafico['X'], y=df_grafico['Y'], hue=df_grafico['Cluster'], palette=sns.color_palette(n_colors=cantidadDeComponentes)).legend().set_visible(False)
    plt.show()
```

# Experimentaciones

### Función para generar dataframes.

```python
# Variable = 'cantidad_de_nodos', 'cantidad_de_clusters' o 'distancia_entre_nodos'.
algoritmos = ['prim', 'kruskal', 'kruskalPC']
def gendf_experimentacion(variable):
    rows = []
    for algoritmo in algoritmos:
        file_name = 'experimentaciones/resultados/%s_%s.out' % (variable, algoritmo)
        lines = [line.rstrip('\n') for line in open(file_name)]
        for i in range(0, len(lines)):
            datos = lines[i].split()
            if len(datos) > 1:
                x = int(datos[0])
                tiempo = int(datos[1])
                rows.append([algoritmo, x, tiempo])
    return pd.DataFrame(rows, columns=['Algoritmo', variable, 'Tiempo (ns)'])
```

### Cantidad de nodos. Generar dataframe y plotear.

```python
variable = 'cantidad_de_nodos'

df_experimentacion = gendf_experimentacion(variable)
plt.figure(figsize=(16, 5))
sns.lineplot(x=df_experimentacion[variable], y=df_experimentacion['Tiempo (ns)'], hue=df_experimentacion['Algoritmo'])
plt.ylabel('Tiempo (decenas de segundo)')
plt.show()

plt.figure(figsize=(16, 5))
sns.lineplot(x=df_experimentacion[variable], y=df_experimentacion['Tiempo (ns)'], hue=df_experimentacion['Algoritmo']).set(yscale = 'log')
plt.show()
```

### Cantidad de clusters. Generar dataframe y plotear.

```python
variable = 'cantidad_de_clusters'
df_experimentacion = gendf_experimentacion(variable)
plt.figure(figsize=(6, 4))
sns.lineplot(x=df_experimentacion[variable], y=df_experimentacion['Tiempo (ns)'], hue=df_experimentacion['Algoritmo'])
plt.ylabel('Tiempo (segundos)')
plt.show()
```

### Distancia entre nodos. Generar dataframe y plotear.

```python
variable = 'distancia_entre_nodos'
df_experimentacion = gendf_experimentacion(variable)
plt.figure(figsize=(6, 4))
sns.lineplot(x=df_experimentacion[variable], y=df_experimentacion['Tiempo (ns)'], hue=df_experimentacion['Algoritmo'])
plt.show()
```

### Variación de f (factor de incosistencia) y d (máxima desviación). Generar dataframes y plotear.

```python
def gendf_grafico(n,f,d):
    rows = []
    file_name = 'tests/problemas/problema_%i.in' % (n)
    results_file_name = 'tests/variar-d-f/kruskalPC %i %i/resultado_%i.out' % (f,d,n)
    lines = [line.rstrip('\n') for line in open(file_name)]
    results_lines = [line.rstrip('\n') for line in open(results_file_name)]
    for i in range(0, len(lines)):
        datos = lines[i].split()
        if len(datos) > 1:
            x = int(datos[0])
            y = int(datos[1])
            componente = int(results_lines[i-1])
            rows.append([x, y, componente])
    return pd.DataFrame(rows, columns=['X', 'Y', 'Cluster'])
```

```python
n = 5
for f in [1,2,15]:
    df_grafico = gendf_grafico(n,f,1)
    cantidadDeComponentes = df_grafico['Cluster'].max(axis=0) + 1
    sns.scatterplot(x=df_grafico['X'], y=df_grafico['Y'], hue=df_grafico['Cluster'], palette=sns.color_palette(n_colors=cantidadDeComponentes)).legend().set_visible(False)
    title = 'Con f = %i y d = 1.' % (f)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel('')
    plt.ylabel('')
    plt.show()

n = 8
for f in [1,2,3,6,15,20]:
    df_grafico = gendf_grafico(n,f,1)
    cantidadDeComponentes = df_grafico['Cluster'].max(axis=0) + 1
    sns.scatterplot(x=df_grafico['X'], y=df_grafico['Y'], hue=df_grafico['Cluster'], palette=sns.color_palette(n_colors=cantidadDeComponentes)).legend().set_visible(False)
    title = 'Con f = %i y d = 1.' % (f)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
```

### Correlación con la complejidad teórica. Generar dataframes y plotear.

```python
variable = 'cantidad_de_nodos'

def gendf_correlacion(algoritmo):
    rows = []
    file_name = 'experimentaciones/resultados/cantidad_de_nodos_%s.out' % (algoritmo)
    lines = [line.rstrip('\n') for line in open(file_name)]
    for i in range(0, len(lines)):
        datos = lines[i].split()
        if len(datos) > 1:
            x = int(datos[0])
            tiempo = int(datos[1])
            if algoritmo == 'prim':
                y = x**3
            elif algoritmo == 'kruskal':
                y = (x**2) * math.log2(x)
            else: 
                y = x**2
            rows.append([algoritmo, x, tiempo, y])
    return pd.DataFrame(rows, columns=['Algoritmo', variable, 'Tiempo (ns)', 'Complejidad teórica'])

for algoritmo in algoritmos:
    df_experimentacion = gendf_correlacion(algoritmo)

    sns.jointplot(data=df_experimentacion, x='Tiempo (ns)', y='Complejidad teórica', kind='reg', color='g')
    if algoritmo == 'prim':
        plt.xlabel('Tiempo del algoritmo de Prim (decenas de segundo)')
        plt.ylabel('Complejidad teórica (n^3)')
        
        r = np.corrcoef(df_experimentacion['Complejidad teórica'], df_experimentacion['Tiempo (ns)'])[0, 1]
        print('El coeficiente de correlación de Pearson en Prim es '+ str(r) + '.')
    elif algoritmo == 'kruskal':
        plt.xlabel('Tiempo del algoritmo de Kruskal (segundos)')
        plt.ylabel('Complejidad teórica ((n^2)*log(n))')
        
        
        r = np.corrcoef(df_experimentacion['Complejidad teórica'], df_experimentacion['Tiempo (ns)'])[0, 1]
        print('El coeficiente de correlación de Pearson en Kruskal es '+ str(r) + '.')
    else:
        plt.xlabel('Tiempo del algoritmo de Kruskal con P.C. (segundos)')
        plt.ylabel('Complejidad teórica ((n^2)*alpha(n))')
        
        
        r = np.corrcoef(df_experimentacion['Complejidad teórica'], df_experimentacion['Tiempo (ns)'])[0, 1]
        print('El coeficiente de correlación de Pearson en Kruskal con P.C. es '+ str(r) + '.')
    plt.yticks([])
```
